```python
import pandas as pd
import numpy as np
import networkx as nx
from typing import Optional, List
import matplotlib.pyplot as plt
from dataset import nasdaq_dataset as dataset
from dataset import edge_creation as edgecreation
```

```python
dataset_var = dataset.NASDAQDataset(base_dir="dataset/", metadata_file="dataset/dataset/symbols_valid_meta.csv")
```

```python
stock_list = ["AAPL", "AMZN","MSM"]
test_date = "2020-01-02"

# Create the daily graph for the test date
daily_graph = edgecreation.create_daily_graph(dataset_var, stock_list, test_date)

# Display nodes with features
print("Nodes and their features:")
for node in daily_graph.nodes(data=True):
    print(node)

# Display edges with weights
print("\nEdges and their weights:")
for edge in daily_graph.edges(data=True):
    print(edge)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(daily_graph)
nx.draw(daily_graph, pos, with_labels=True, node_size=700, node_color="lightblue", font_size=10)
edge_labels = nx.get_edge_attributes(daily_graph, 'weight')
nx.draw_networkx_edge_labels(daily_graph, pos, edge_labels=edge_labels)
plt.title(f"Daily Correlation Graph for {test_date}")
plt.show()
```
